```python
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df['Voice mail plan'].value_counts()
```

```python
df.loc[df['Voice mail plan'] == 'No', 'Voice mail plan'] = 0
df.loc[df['Voice mail plan'] == 'Yes', 'Voice mail plan'] = 1
```

```python
df['Area code'].value_counts()
```

```python
le = LabelEncoder()
df["State"] = le.fit_transform(df["State"].astype(str)) 
```

```python
df.drop(["Area code"], axis = 1, inplace = True)
```

```python
df['Number vmail messages'].value_counts()
```

```python
df.drop(['Number vmail messages'], axis = 1, inplace = True)
```

```python
train=df.drop('Churn', axis=1)
df
```

```python
X = train.values.reshape(-1, 1)
Y = df['Churn'].values.reshape(-1, 1)  
regression = LogisticRegression()  
regression.fit(X, Y)
Y_pred = regression.predict(X)  
regression.score(Y, Y_pred)
```
